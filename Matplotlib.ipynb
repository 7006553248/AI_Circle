```python
import numpy as np
import matplotlib.pyplot as plt

X = np.linspace(0.0, 4*np.pi,100)
print 'X:',X,'\n'
y = np.linspace(0.0,14.0,15)
print 'Y:',y,'\n'
C,S = np.cos(X), np.sin(X)
z = np.zeros(15)


plt.plot(X,C)
plt.plot(X,S)
plt.plot(y,z,c='r')

plt.xlabel('X')
plt.ylabel('Y')

plt.show()
```

```python
n = 14
X = np.arange(n)
print 'X: ',X,'\n'
Y1 = (1-X/float(n)) * np.random.uniform(0.5,1.0,n)
print "Y1: ",Y1,'\n'


Y2 = -(X/float(n)) * np.random.uniform(0.5,1.0,n)
print "Y1: ",Y1,'\n'

plt.bar(X, Y1, facecolor='b', edgecolor='r')
plt.bar(X, Y2, facecolor='r', edgecolor='blue')

for x,y in zip(X,Y1):
    plt.text(x+0.4, y+0.05, '%.2f' % y, ha='center')
    
for x,y in zip(X,Y2):
    plt.text(x+0.4, y-0.12, '%.2f' % y, ha='center')


plt.ylim(-1,+1)


plt.show()
```

```python
x = np.linspace(0, 2 * np.pi, 400)
y = np.sin(x ** 2)

f, (ax1, ax2, ax3) = plt.subplots(3, sharex=True, sharey=True)
ax1.plot(x, y)
ax1.set_title('Sharing both axes')
ax2.scatter(x, y)
ax3.scatter(x, 2 * y ** 2 - 1, color='r')


f.subplots_adjust(hspace=0)
#plt.setp([a.get_xticklabels() for a in f.axes[:-1]], visible=False)

plt.show()
```
