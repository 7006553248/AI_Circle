```python
import numpy as np
import matplotlib.pyplot as plt

X = np.linspace(-2*np.pi, 2*np.pi,200)
print X
x1 = np.linspace(-2*np.pi, 2*np.pi,2)
print 'x1:',x1,'\n'
C,S = np.cos(X), np.sin(X)
xasis = np.zeros(2)


plt.plot(X,C,c='#f4df42',alpha=.5)
plt.plot(X,S)
plt.plot(x1,xasis,c='black',alpha=.5)

plt.xlabel('XX')
plt.ylabel('House Price')

plt.xticks(np.linspace(-2*np.pi,2*np.pi,5))
plt.yticks([-1,0,1])

plt.show()
```

```python
n = 14
X = np.arange(n)
print 'X: ',X,'\n'
Y1 = (1-X/float(n)) * np.random.uniform(0.5,1.0,n)
print "Y1: ",Y1,'\n'


Y2 = -(X/float(n)) * np.random.uniform(0.5,1.0,n)
print "Y2: ",Y2,'\n'

plt.bar(X, Y1, facecolor='b', edgecolor='r')
plt.bar(X, Y2, facecolor='r', edgecolor='blue')

for x,y in zip(X,Y1):
    plt.text(x+0.4, y+0.05, '%.2f' % y, ha='center')
    
for x,y in zip(X,Y2):
    plt.text(x+0.4, y-0.12, '%.2f' % y, ha='center')


plt.ylim(-1,+1)


plt.show()
```

```python
x = np.linspace(0, 2 * np.pi, 50)
y = np.sin(x)

f, (ax1, ax2, ax3) = plt.subplots(3)
ax1.plot(x, y)
ax1.set_title('Sharing both axes')
ax2.scatter(x, y)
ax3.scatter(x, 2 * y ** 2 - 1, color='r')


f.subplots_adjust(hspace=0)
#plt.setp([a.get_xticklabels() for a in f.axes[:-1]], visible=False)

plt.show()
```

```python
x = np.linspace(0, 2 * np.pi, 20)
y = np.sin(x)

f, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2,sharex='col', sharey='row' )
ax1.plot(x, y)
ax1.set_title('Sharing x per column, y per row\n')
ax2.scatter(x, y)
ax3.scatter(x, 2 * y ** 2 - 1, color='r')
ax4.plot(x, 2 * y ** 2 - 1, color='r')
f.subplots_adjust(hspace=0)
plt.show()
```
