```python
import numpy as np
import matplotlib.pyplot as plt

X = np.linspace(0.0, 4*np.pi,100)
print 'X:',X,'\n'
y = np.linspace(0.0,14.0,15)
print 'Y:',y,'\n'
C,S = np.cos(X), np.sin(X)
z = np.zeros(15)


plt.plot(X,C)
plt.plot(X,S)
plt.plot(y,z,c='r')

plt.show()
```
